Unsupervised Learning: K-Means Clustering

Synthetic Data generated

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
# Step 1: Generate synthetic dataset
n_samples = 300
n_features = 2
n_clusters = 3
random_state = 42
```

```python
data, labels_true = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=random_state)

# Standardize the dataset
data = StandardScaler().fit_transform(data)
```

```python
# Step 2: Visualize the dataset
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], s=50, alpha=0.6, edgecolors='k')
plt.title("Generated Synthetic Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Step 3: Apply KMeans clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=random_state, n_init=10)
kmeans.fit(data)
labels_pred = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Step 4: Visualizing cluster assignments
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=labels_pred, cmap='viridis', s=50, alpha=0.6, edgecolors='k')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("KMeans Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
# Step 5: Interpretation
print("Cluster Centers:\n", centroids)
print("Cluster Labels:\n", np.unique(labels_pred))
```

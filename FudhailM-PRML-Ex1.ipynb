```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
url = "/Users/fudhail/Downloads/advertising.csv"
advertising = pd.read_csv(url)
```

```python
# Display first few rows
advertising.head()
```

```python
# Basic Info
advertising.info()

# Summary statistics
advertising.describe()
```

```python
# Check for missing values
print("Missing values per column:")
print(advertising.isnull().sum())
```

```python
# Distribution of Sales
plt.figure(figsize=(8, 5))
sns.histplot(advertising['Sales'], bins=20, kde=True, color='blue')
plt.title("Distribution of Sales")
plt.show()
```

```python
# Pairplot to visualize relationships
sns.pairplot(advertising)
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(advertising.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Scatter plots of Sales vs Advertising Budgets
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
sns.scatterplot(x='TV', y='Sales', data=advertising, ax=axes[0], color='red')
axes[0].set_title("Sales vs TV Budget")
sns.scatterplot(x='Radio', y='Sales', data=advertising, ax=axes[1], color='green')
axes[1].set_title("Sales vs Radio Budget")
sns.scatterplot(x='Newspaper', y='Sales', data=advertising, ax=axes[2], color='blue')
axes[2].set_title("Sales vs Newspaper Budget")
plt.show()
```

```python
# Boxplots to check outliers
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
sns.boxplot(y=advertising['TV'], ax=axes[0], color='red')
axes[0].set_title("TV Budget")
sns.boxplot(y=advertising['Radio'], ax=axes[1], color='green')
axes[1].set_title("Radio Budget")
sns.boxplot(y=advertising['Newspaper'], ax=axes[2], color='blue')
axes[2].set_title("Newspaper Budget")
plt.show()
```
